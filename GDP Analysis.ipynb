```python
#Part I-B:
import pandas as pd
import numpy as np
import glob,os
```

```python
#Concatenating all the files of  different states into a dataframe.
path = r'C:\Users\Narendra Raushan\Downloads\GDP Assignment Files\Indian States'
all_files=glob.glob(path + "/*.csv")
li=[]
for filename in all_files:
    df = pd.read_csv(filename,encoding='unicode_escape',index_col=None,header=0)
    df=df[['Item','2014-15']]
    df=df.T
    new_header=df.iloc[0]
    df=df[1:]
    df.columns=new_header
    df['State']=filename.split('-')[1]   
    li.append(df)
frame=pd.concat(li,axis=0,ignore_index=True,sort=True)
frame
    
```

```python
frame.columns
```

```python
frame
```

```python
#Setting state as index.
frame.set_index(['State'],inplace=True)
```

```python
#Selecting the data.
frame.loc[:,'Per Capita GSDP (Rs.)']
```

```python
#Sorting the values based on Per capita GSDP and plotting the graph.
c=frame.sort_values(by='Per Capita GSDP (Rs.)',ascending=False)
c
frame=frame.sort_values(by='Per Capita GSDP (Rs.)',ascending=False)
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(15,10)
frame[['Per Capita GSDP (Rs.)']].plot.bar()
```

```python
#Perecentage contribution of each sector.
d=c.loc[:, ['Primary', 'Secondary', 'Tertiary','Gross State Domestic Product','Primary%','Secondary%','Tertiary%']]
c['Primary%']   = c['Primary'] / c['Gross State Domestic Product']*100
c['Secondary%'] = c['Secondary'] / c['Gross State Domestic Product']*100
c['Tertiary%']  = c['Tertiary'] / c['Gross State Domestic Product']*100
d
```

```python
#Sorting by GSDP 
e=d.sort_values(by='Gross State Domestic Product',ascending=False)
e
```

```python
#Plotting percentage contibution of each sector.
import matplotlib.pyplot as plt
e[['Primary%','Secondary%','Tertiary%']].plot(kind='bar', stacked=True)
```

```python
#Resetting index.
d=c.reset_index()
#Dividing the states into quantiles based on per capita GSDP.
d['Per Capita GSDP (Rs.)'].quantile([0.20,0.5,0.85,1])
d
d.set_index(['State'],inplace=True)
```

```python
#Inserting the quantile value for different states into different categories based on the Per capita GSDP.
d['Category']=pd.qcut(d.loc['Goa':'Bihar','Per Capita GSDP (Rs.)'],[.0,.20,.5,.85,1],labels=['C1','C2','C3','C4'])
d
```

```python
#Removal of sub-sub sectors.
e=d.drop(['Crops','Livestock','Forestry and logging','Fishing and aquaculture','Trade & repair services','Hotels & restaurants','Road transport','Road transport*','Road transport**','Railways','Water transport','Air transport','Services incidental to transport', 'Services incidental to transport*','Storage','Primary%', 'Secondary%', 'Tertiary%','Primary','Secondary','Tertiary','Trade & repair services*'],axis=1)
e
```

```python
#Summing of sub sectors by each category of states.
f=e.groupby('Category').sum()
f
```

```python
#Ratio of contribution of different sub sectors to the GSDP.
f['Agriculture, forestry and fishing %'] = f['Agriculture, forestry and fishing'] / f['Gross State Domestic Product'] * 100
f['Mining and quarrying %'] = f['Mining and quarrying'] / f['Gross State Domestic Product'] * 100
f['Manufacturing %'] = f['Manufacturing'] / f['Gross State Domestic Product'] * 100
f['Electricity, gas, water supply & other utility services %'] = f['Electricity, gas, water supply & other utility services'] / f['Gross State Domestic Product'] * 100
f['Construction %'] =f['Construction'] / f['Gross State Domestic Product'] * 100
f['Trade, repair, hotels and restaurants %'] = f['Trade, repair, hotels and restaurants'] / f['Gross State Domestic Product'] * 100
f['Transport, storage, communication & services related to broadcasting %'] = f['Transport, storage, communication & services related to broadcasting'] / f['Gross State Domestic Product'] * 100
f['Financial services %'] = f['Financial services'] / f['Gross State Domestic Product'] * 100
f['Real estate, ownership of dwelling & professional services %'] = f['Real estate, ownership of dwelling & professional services'] / f['Gross State Domestic Product'] * 100
f['Public administration %'] = f['Public administration'] / f['Gross State Domestic Product'] * 100
f['Other services %'] = f['Other services'] / f['Gross State Domestic Product'] * 100
```
