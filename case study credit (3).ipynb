The outliers in some of the above columns have been treated as below

```python
# Calculating the quartiles for the column 'CNT_CHILDREN' to find the outliers
working_df['CNT_CHILDREN'].quantile([0.25,0.5,0.75,0.9,0.95,0.98,0.99,0.9999,0.99999,1])
```

```python
working_df['CNT_CHILDREN'] = np.where(working_df['CNT_CHILDREN']>3, 3, working_df['CNT_CHILDREN'])
```

```python
# Inspecting the column after treatment of outliers by capping
working_df.boxplot(['CNT_CHILDREN'])
plt.show()
```

The above boxplot still shows '3' as an outlier. However, there cannot be any values between 2 and 3 for 'CNT_CHILDREN' and this value is very much possible and hence, is not being capped/ replaced.
